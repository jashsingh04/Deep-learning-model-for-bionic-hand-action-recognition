```python
# Function to split each column into rows of 1000 values
def split_into_rows(column, sensor_name):
    num_rows = len(column) // 1000
    data_rows = [column[i * 1000: (i + 1) * 1000] for i in range(num_rows)]
    return [(sensor_name, row) for row in data_rows]

# Create a new DataFrame with data in the desired format
new_data = []
for col_name in df.columns:
    sensor_name = col_name
    sensor_values = df[col_name].values
    new_data.extend(split_into_rows(sensor_values, sensor_name))

result_df = pd.DataFrame(new_data, columns=['sensor_name', 'sensor_data'])

# Split 'sensor_data' into respective columns
result_df = pd.DataFrame(result_df['sensor_data'].tolist(), index=result_df['sensor_name']).reset_index()

# Print the resulting DataFrame
result_df.rename(columns={'sensor_name': 'EMG'}, inplace=True)
result_df = result_df.assign(label='cylindrical')

result_df
```

```python
file_names = [
    'SumantRH07072023_hook.xlsx',
    'SumantRH07072023_Lateral.xlsx',
    'SumantRH07072023_Palmar.xlsx',
    'SumantRH07072023_Spherical.xlsx',
    'SumantRH07072023_Tip.xlsx'
]
frames = [result_df]
for file in file_names:
    dh = pd.read_excel(f"D:/project/DRDO/ML_gui_v2/Data Collection 07072023/Sumant/RH/{file}")
    file_name_without_extension = file.split('.')[0]  # Remove the file extension (".xlsx")
    action_name = file_name_without_extension.split('_')[-1]
    # Function to split each column into rows of 1000 values
    def split_into_rows(column, sensor_name):
        num_rows = len(column) // 1000
        data_rows = [column[i * 1000: (i + 1) * 1000] for i in range(num_rows)]
        return [(sensor_name, row) for row in data_rows]

    # Create a new DataFrame with data in the desired format
    new_data = []
    for col_name in dh.columns:
        sensor_name = col_name
        sensor_values = dh[col_name].values
        new_data.extend(split_into_rows(sensor_values, sensor_name))

    result_df_new = pd.DataFrame(new_data, columns=['sensor_name', 'sensor_data'])

    # Split 'sensor_data' into respective columns
    result_df_new = pd.DataFrame(result_df_new['sensor_data'].tolist(), index=result_df_new['sensor_name']).reset_index()

    # Print the resulting DataFrame
    result_df_new.rename(columns={'sensor_name': 'EMG'}, inplace=True)
    result_df_new = result_df_new.assign(label=action_name)
    frames.append(result_df_new)
    print(result_df_new)

combined_data = pd.concat(frames)  
combined_data
```

```python
combined_data['EMG'] = combined_data['EMG'].str.replace(' ','')
combined_data.to_csv('Sumant_RH.csv')
```
